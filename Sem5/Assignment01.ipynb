```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv("/content/drive/MyDrive/ColabNotebooks/biostats.csv")
```

```python
age_array = ds["Age"].values
```

```python
print("Age column:")
print(age_array)
```

```python
average_age = age_array.mean()
std_dev_age = age_array.std()
min_age = age_array.min()
max_age = age_array.max()
print("\nStatistics for Age column:")
print(f"Average Age: {average_age:.2f}")
print(f"Standard Deviation: {std_dev_age:.2f}")
print(f"Minimum Age: {min_age}")
print(f"Maximum Age: {max_age}")
```

```python
percentiles = [30, 50, 70]
age_percentiles = [np.percentile(age_array, p) for p in percentiles]
print("\nPercentiles for Age column:")
for p, value in zip(percentiles, age_percentiles):
    print(f"{p}th Percentile: {value}")
```

```python
plt.figure(figsize=(8, 6))
plt.hist(age_array, bins=20, color='blue', edgecolor='red')
plt.title("Age Distribution – Vedantsinh")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
