```python
# Checking for potential overfitting

from keras import models
from keras import layers
from keras.datasets import fashion_mnist
from sklearn.model_selection import train_test_split

# Load the Fashion MNIST dataset
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

# Reshape and normalize the images
train_images_normalized = train_images / 255.0
train_images_normalized = train_images_normalized.reshape((-1, 28, 28))

# Split the data into training and validation sets
train_images, val_images, train_labels, val_labels = train_test_split(
    train_images_normalized, train_labels, test_size=0.2, random_state=42
)

# Define the model
myNetwork = models.Sequential()
myNetwork.add(layers.Reshape((784,), input_shape=(28, 28)))
myNetwork.add(layers.Dense(512, activation='relu'))
myNetwork.add(layers.Dense(128, activation='relu'))
myNetwork.add(layers.Dense(10, activation='softmax'))

# Compile the model
myNetwork.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
myNetwork.fit(train_images, train_labels, epochs=10, validation_data=(val_images, val_labels))

# Evaluate the model on the test set
test_images_normalized = test_images / 255.0
test_images_normalized = test_images_normalized.reshape((-1, 28, 28))
test_loss, test_acc = myNetwork.evaluate(test_images_normalized, test_labels)

# Print the achieved accuracy and loss on the test set
print("Test Accuracy:", test_acc)
print("Test Loss:", test_loss)

# Check for overfitting by comparing training and validation accuracies
train_loss, train_acc = myNetwork.evaluate(train_images, train_labels)
val_loss, val_acc = myNetwork.evaluate(val_images, val_labels)

print("Training Accuracy:", train_acc)
print("Validation Accuracy:", val_acc)

# Check for overfitting
if train_acc > val_acc:
    print("Potential overfitting: Training accuracy is higher than validation accuracy.")
else:
    print("No significant overfitting observed.")
```
